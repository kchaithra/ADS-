```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
```

```python

# Load the graph from karate.gml file
G = nx.read_gml('karate.gml')

# Iterate through each node and print its metadata
for node in G.nodes():
    metadata = G.nodes[node]
    print(f"Node {node}: {metadata}")
```

```python
# iii. Calculate all types of centrality (degree, betweenness, closeness, eigenvector, pagerank centrality). Analyse the nodes based on the centrality values.
degree_cent = nx.degree_centrality(G)
# Sort nodes based on centrality values
sorted_degree_cent = sorted(degree_cent.items(), key=lambda x: x[1], reverse=True)
# Print top nodes for each centrality measure
print("Degree Centrality:")
print(sorted_degree_cent[:5])
nx.draw_networkx(G, node_size=[2000 * v for v in degree_cent.values()])
```

```python
betweenness_cent = nx.betweenness_centrality(G)
sorted_betweenness_cent = sorted(betweenness_cent.items(), key=lambda x: x[1], reverse=True)
print("Betweenness Centrality:")
print(sorted_betweenness_cent[:5])
nx.draw_networkx(G, node_size=[2000 * v for v in betweenness_cent.values()])
```

```python
closeness_cent = nx.closeness_centrality(G)
sorted_closeness_cent = sorted(closeness_cent.items(), key=lambda x: x[1], reverse=True)
print("Closeness Centrality:")
print(sorted_closeness_cent[:5])
nx.draw_networkx(G, node_size=[1000 * v for v in closeness_cent.values()])
```

```python
eigenvector_cent = nx.eigenvector_centrality(G)
sorted_eigenvector_cent = sorted(eigenvector_cent.items(), key=lambda x: x[1], reverse=True)
print("Eigenvector Centrality:")
print(sorted_eigenvector_cent[:5])
nx.draw_networkx(G, node_size=[2400 * v for v in eigenvector_cent.values()])
```

```python
pagerank_cent = nx.pagerank(G)
sorted_pagerank_cent = sorted(pagerank_cent.items(), key=lambda x: x[1], reverse=True)
print("Pagerank Centrality:")
print(sorted_pagerank_cent[:5])
nx.draw_networkx(G, node_size=[5000 * v for v in pagerank_cent.values()])
```

```python
# iv. Find the possible k-components of the network and the clustering coefficient.
components = nx.k_components(G)
print("K-components:")
print(components)
```

```python
clustering_coefficient = nx.clustering(G)
print("Clustering Coefficient:", clustering_coefficient)
```

```python
ave_coefficient = nx.average_clustering(G)
print(" average Clustering Coefficient:", ave_coefficient)
```

```python
# v. Find out the communities using Girvan-Newman algorithm and Louvain method.
# Girvan-Newman algorithm
communities_girvan_newman = community.girvan_newman(G)
print("Communities (Girvan-Newman):", tuple(sorted(c) for c in next(communities_girvan_newman)))

from networkx.algorithms import community
comp = community.girvan_newman(G)
communities = tuple(sorted(c) for c in next(comp))
len(communities)
```

```python
com1 = G.subgraph(communities[0])
com2 = G.subgraph(communities[1])
```

```python
nx.draw_networkx(com1)
```

```python
nx.draw_networkx(com2)
```

```python
# Louvain method (Note: You need to install python-louvain library)
import community as community_louvain
communities_louvain = community_louvain.best_partition(G)
print("Communities (Louvain):", communities_louvain)
louvian_comms = community.louvain_communities(G)
len(louvian_comms)
```

```python
lcom1 = G.subgraph(louvian_comms[0])
lcom2 = G.subgraph(louvian_comms[1])
lcom3 = G.subgraph(louvian_comms[2])
lcom4 = G.subgraph(louvian_comms[3])
```
